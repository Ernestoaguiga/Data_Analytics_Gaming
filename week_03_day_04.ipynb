```python
# La orden mágica en SQL
query = """
SELECT Name, Genre, Global_Sales
FROM Ventas_Videojuegos
WHERE Global_Sales > 30
"""

# Le pedimos a pandas que nos muestre el resultado de la orden
resultado = pd.read_sql_query(query, conexion)
display(resultado)
```

```python
# Queremos el Nombre y las ventas en Japón
# Los ordenamos por JP_Sales de forma Descendente (el más alto primero)
# Y nos quedamos solo con los 5 mejores
query_japon = """
SELECT Name, JP_Sales
FROM Ventas_Videojuegos
ORDER BY JP_Sales DESC
LIMIT 5
"""

resultado_japon = pd.read_sql_query(query_japon, conexion)
display(resultado_japon)
```

```python
# Queremos el Género y la SUMA de sus ventas globales
# Le ponemos el nombre 'Ventas_Totales' a esa suma
# Agrupamos por Género
# Y ordenamos para ver quién es el más rentable de todos
query_agrupada = """
SELECT Genre, SUM(Global_Sales) AS Ventas_Totales
FROM Ventas_Videojuegos
GROUP BY Genre
ORDER BY Ventas_Totales DESC
"""

resultado_pro = pd.read_sql_query(query_agrupada, conexion)
display(resultado_pro)
```
